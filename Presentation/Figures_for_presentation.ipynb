```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(123)

X = np.random.multivariate_normal(mean =np.ones(2), cov = np.array([[1, 0.5], [0.5, 1]]), size = 200)

error1= np.random.normal(0, 1, 100)
error2= np.random.normal(0, 1, 100)

beta1 = np.array([1, 1])
beta2 = np.array([-1, -1])

Y1 = X[0:100, ]@beta1 + error1
Y2 = X[100:200, ]@beta2 + error2

Y = np.r_[Y1, Y2]

time1 = np.arange(start=0, stop=100, step=1)
time2 = np.arange(start=100, stop=200, step=1)

plt.figure(figsize  = (8, 4))
plt.plot(time1, Y1, label = r'$y_t = X_t\beta_1 + \epsilon_t$')
plt.plot(time2, Y2, label = r'$y_t = X_t\beta_2 + \epsilon_t$')
plt.axvline(x = 100, color = 'gray', ls = "dashed")
plt.text(x = 105, y = 4,  s = "Change point",color = "black")
plt.xlabel("time")
plt.ylabel(r'$y_t$')
plt.legend()
```
